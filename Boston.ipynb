```python
import os

import pandas as pd



data = pd.read_csv(r'C:\Study\Code\数据集\Bosch-Engine-Datasets-master\gengine1\raw_000010_measurement_000000.csv').values
file_path = "C:\Study\Code\数据集\Bosch-Engine-Datasets-master\gengine1"
file_names = os.listdir(r'C:\Study\Code\数据集\Bosch-Engine-Datasets-master\gengine1')
train_index = file_names[:20]
test_index = file_names[20:]
```

```python

train_data = pd.DataFrame([])
for i in train_index:
    d = pd.read_csv(fr'{file_path}\{i}',names = ['Spped','Load','Lambada','Ignition Angle','Fuel Cutoff'])
    print(d.shape)
    train_data =  pd.concat([train_data,d],axis=0,ignore_index=True)
print(train_data.shape)
train_data.describe()
```

```python

test_data = pd.DataFrame([])
for i in test_index:
    d = pd.read_csv(fr'{file_path}\{i}',names = ['Spped','Load','Lambada','Ignition Angle','Fuel Cutoff'])
    print(d.shape)
    test_data =  pd.concat([test_data,d],axis=0,ignore_index=True)
print(test_data.shape)
test_data.describe()
```
